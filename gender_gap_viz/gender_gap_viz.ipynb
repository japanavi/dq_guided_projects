# Visualizing The Gender Gap in College Degrees
The goal of this notebook is to see if there is a gender gap in college degrees, primarily in Science Technology Engineering and Mathematics (STEM) related degrees. To accomplish this, multiple visualization techniques will be implemented to see if the number of college degrees awarded to women in STEM fields each year are significantly lower than the number awarded to men. 

## Resources:
* Supplied data is from [The Department of Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp)
* Data cleaned and prepared by [Randal Olson](http://www.randalolson.com/)
    * [Download data](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv)

### Initial visualization (code supplied)

```python
import pandas as pd 
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

### Degree categories

```python
stem_cats = [
    'Psychology', 
    'Biology', 
    'Math and Statistics', 
    'Physical Sciences', 
    'Computer Science', 
    'Engineering']

lib_arts_cats = [
    'Foreign Languages', 
    'English', 
    'Communications and Journalism', 
    'Art and Performance', 
    'Social Sciences and History']

other_cats = ['Health Professions', 
              'Public Administration', 
              'Education', 
              'Agriculture',
              'Business', 
              'Architecture']
```

### Visualizing with subplots

```python
# Helper function 
def plot_data(axis, column):
    """Plots data from women_degreese[column] on axis"""
    axis.plot(women_degrees['Year'], women_degrees[column], c=cb_dark_blue, linewidth=3) 
    axis.plot(women_degrees['Year'], 100 - women_degrees[column], c=cb_orange, linewidth=3)
    axis.set_title(f'{column}')
    axis.tick_params(axis="both", bottom=False, left=False)
    axis.set_xlim(1968, 2011)
    axis.set_ylim(0,100)
    
# Creating figure
fig = plt.figure(figsize=(16, 20))

# First column (stem)
for i, stem_major in enumerate(stem_cats):
    ax = fig.add_subplot(6, 3, i*3+1, frame_on=False)
    plot_data(ax, stem_major)
    
    if i == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif i == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
    
# Second column (liberal arts)
for i, lib_arts_major in enumerate(lib_arts_cats):
    ax = fig.add_subplot(6, 3, i*3+2, frame_on=False)
    plot_data(ax, lib_arts_major)
    
    if i == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    
# Third column (other)
for i, other_major in enumerate(other_cats):
    ax = fig.add_subplot(6, 3, i*3+3, frame_on=False)
    plot_data(ax, other_major)
    
    if i == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif i == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')

plt.tight_layout()
plt.show();
```

# Conclusion 
It is very obvious from the figures above that the percentage of degrees awarded to men and women is staggeringly different in most fields. However, in the cases of Social Sciences and History, Math and Statistics, Agriculture, and Business, there is almost a 50-50 split. As for the main goal of this analysis, it is clear that the percentage of women graduating with a degree in a STEM field is very low compared to men, although in some areas such as psychology and biology, the percent of women graduating with these STEM degrees is higher than men. Despite this fact, out of the six major STEM degree fields, women are outnumbered by men in all but two. From this I conclude that there was (and may still be) a gender gap in STEM degrees in the period from 1970 - 2010. However, it does appear that the percentage of women in most of the STEM fields is steadily increasing.
